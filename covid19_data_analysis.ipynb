```python
df=corona.drop(['Lat','Long'],axis=1)
df.head(10)
```

```python
countries=list(df_agre.index)
max_infection_rate= [] #()tupole #{}dict
for c in countries:

  max_infection_rate.append(df_agre.loc[c].diff().max())
```

```python
max_infection_rate
```

```python
df_agre['max_infection_rate']=max_infection_rate
```

```python
corona_data=pd.DataFrame(df_agre['max_infection_rate'])
```
